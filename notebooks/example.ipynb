```python
%matplotlib inline

from microscopestitching import stitch
from skimage.io import imsave, imshow
```

```python
!ls r*png
```

```python
def img_tuple(img):
    row = int(img[1])
    col = int(img[4])
    return img, row, col
```

```python
files = !ls r*png
images = [img_tuple(f) for f in files]
# path, row, column
images[:3]
```

```python
merged = stitch(images)
```

```python
merged.shape, merged.dtype
```

```python
merged.min(), merged.max()
```

```python
imsave('merged.png', merged)
imshow(merged);
```
